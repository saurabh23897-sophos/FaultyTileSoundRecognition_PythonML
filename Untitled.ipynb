```python
import os
import sys
import cv2
import time
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_files
from keras.utils import np_utils
from keras.models import Sequential, Model, load_model, model_from_json
```

```python
# load json and create model
num = 1
json_file = open("resnet_model_m" + str(num) + ".json", "r")
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("resnet_model_w" + str(num) + ".h5")
print("Loaded model from disk")
```

```python
print(loaded_model.summary())
```

```python
imgs = np.empty((1000,3,50,100), dtype='uint8')
t = time.time()
for i in range(0,imgs.shape[0]):
    img = cv2.imread('F:/IIT/Sem 7/BTP/Data/Final/16kData/2000ms/Correct/'+str(i+10000)+'.png')
    img = cv2.resize(img, (100,50))
    imgs[i, ...] = img.transpose(2, 0, 1)
    if(i%100==1):
        elapsed = time.time() - t
        print(str(i)+' -> '+str(elapsed))
imgs = imgs.transpose(0, 2, 3, 1)
imgs = imgs/255.
```

```python
y_pred = loaded_model.predict(imgs, verbose=1)
```
